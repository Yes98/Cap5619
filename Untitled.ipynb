```python
from keras.models import Sequential
from keras import layers
from keras.layers import Dense, GRU, Activation
import keras
import keras.utils as ku
from keras.preprocessing import sequence
import numpy as np
import pandas as pd
import time
```

```python
model = Sequential()

model.add(layers.Dense(1000,activation='relu',input_shape=(100,3)))#,input_shape=(None,300)))
#model.add(layers.Dense(1000,activation='relu'))
#model.add(layers.Dense(500,activation='relu'))

model.add(layers.LSTM(256))#, return_sequences=True))
model.add(layers.Dropout(0.2))
#model.add(layers.Dropout(0.4))         
model.add(layers.Dense(100,activation="relu"))
```

```python
#sgd = keras.optimizers.SGD(lr = 0.01, momentum=0.9)
#model.compile(loss='mean_squared_error', optimizer=sgd)
model.compile(loss='mean_squared_error',optimizer='rmsprop',metrics=["accuracy"])
#model.summary()
```

```python

for e in range(5):
    print(f"epoch: {e}")
    startTime = time.perf_counter()
    for z in range(100):
        j = 0
        while True:
            try:
                target = pd.read_csv(f"./VirusData3/sample{z}/Virus{j+1}.csv")
                d = pd.read_csv(f"./VirusData3/sample{z}/Virus{j}.csv")
                x = np.array(d['xcor'])
                x0 = np.zeros((100))
                y = np.array(d['ycor'])
                y0 = np.zeros((100))
                target = np.array(target['sick?']).astype('float').reshape(1,100)
                #print(np.shape(target))
                s = np.array(d['sick?']).astype('float')#.reshape(1,100)
                for i in range(100):
                    try:      
                        if np.isnan(x[i]):
                            pass
                        else:
                            x0[i] = float(x[i])/17
                    except:
                        pass
                    try:
                        if np.isnan(y[i]):
                            pass
                        else:
                            y0[i] = float(y[i])/17
                    except:
                        pass

                    if np.isnan(s[i]):
                        s[i] = 0
                    if np.isnan(target[0,i]):
                        target[0,i] = 0

                #inputs = np.concatenate((x0,y0,s)).reshape(300,)
                #print(target)
                inputs = np.zeros((1,100,3))
                inputs[0,:,0] = x0
                inputs[0,:,1] = y0
                inputs[0,:,2] = s
                #inputs = inputs.reshape(1,100,3)
                #print(inputs)
                #print(target,i)
                hist = model.train_on_batch(inputs,target,reset_metrics=False)
                #print(j)
                #print(hist)
                j += 1
            except:
                
                print(z)
                break
    print(f"Run Time: {time.perf_counter() - startTime}")
    print(model.metrics_names)
    print(hist)

model.summary()
```
